Import Library yang dibutuhkan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

```

Baca file csv yang diperlukan

```python
rs_df = pd.read_csv('21-22 RS.csv',encoding='windows-1252',sep=';')
rs_df
```

Fitur Explanation  
  
1. Player   = Nama Pemain
2. Pos      = Posisi utama Pemain
3. Age      = Usia Pemain
4. Tm       = Nama Tim
5. G        = Game yang sudah dimainkan
6. GS       = Game dimana Pemain menjadi starting line-up
7. MP       = Menit bermain per Game
8. FG       = Jumlah basket dimasukkan per Game (Tanpa Free Throw)
9. FGA      = Jumlah percobaan memasukkan basket per Game (Tanpa Free Throw)
10. FG%     = Persentase Field goal (Tanpa Free Throw)
11. 3P      = Jumlah tembakan 3Pts per Game
12. 3PA     = Jumlah percobaan 3Pts per Game
13. 3P%     = Persentase 3Pts
14. 2P      = Jumlah tembakan 2Pts per Game
15. 2PA     = Jumlah percobaan 2Pts per Game
16. 2P%     = Persentase 2Pts
17. eFG%    = Efisiensi Field Goal
18. FT      = Jumlah Free Throw berhasil per Game
19. FTA     = Jumlah kesempatan Free Throw per Game
20. FT%     = Persentase Free Throw
21. ORB     = Jumlah Rebound Ofensif (di basket lawan) per Game
22. DRB     = Jumlah Rebound Defensif (di basket sendiri) per Game
23. TRB     = Total Rebound
24. AST     = Assist (Operan tepat sebelum Field Goal) per Game
25. STL     = Jumlah pemain berhasil merebut bola dari penguasaan lawan per Game
26. TOV     = Jumlah pemain kehilangan penguasaan bola per Game
27. PF      = Jumlah pemain melakukan pelanggaran per Game
28. PTS     = Jumlah poin yang dicetak pemain per Game

menghapus kolom yang tidak diperlukan

```python
rs_df.drop('Rk', axis=1, inplace=True)
```

Menambahkan Data yang diperlukan

```python
rs_df['MIN'] = rs_df['G'] * rs_df['MP'].round()
rs_df['TFG'] = rs_df['G'] * rs_df['FG'].round()
rs_df['TFGA'] = rs_df['G'] * rs_df['FGA'].round()
rs_df['3PM'] = rs_df['G'] * rs_df['3P'].round()
rs_df['T3PA'] = rs_df['G'] * rs_df['3PA'].round()
rs_df['TAST'] = rs_df['G'] * rs_df['AST'].round()
rs_df['TPTS'] = rs_df['G'] * rs_df['PTS'].round()
rs_df['TORB'] = rs_df['G'] * rs_df['ORB'].round()
rs_df['TDRB'] = rs_df['G'] * rs_df['DRB'].round()
```

# Most Efficient 3Pts Shooters in NBA 2021-22 Regular Season

```python
rs_df[['Player','3P','3PM','3PA','T3PA','3P%']].sort_values(by=['3PM'], ascending=False).head(10)
```

```python
np.random.seed(seed=43)

min_3_attempts = rs_df['T3PA'].min()
max_3_attempts = rs_df['T3PA'].max()

top_5perc_3ptperc = 0.425
shooters_attempts = 200

plt.scatter(x=rs_df['T3PA'], 
            y=rs_df['3P%'], 
            s=(rs_df['3P']**2)*7,
            alpha=.6)
plt.xlabel("Three-points Attempted")
plt.ylabel("Three-points percentage")
plt.title("Best 3 Points shooters in 2021-22 Regular Season");

plt.hlines(y=.37, 
           xmin=min_3_attempts, xmax=max_3_attempts, 
           linestyle='dashed', colors='black');
plt.text(x=max_3_attempts*1.06, 
         y=top_5perc_3ptperc*0.8, 
         s="above 0.95 of players")

f1 = rs_df['3P%'] > top_5perc_3ptperc
f2 = rs_df['T3PA'] > shooters_attempts

df_good_shooters = rs_df[f1 & f2]
steph_curry = rs_df[rs_df['Player'] == 'Stephen Curry']

for i, row in df_good_shooters.iterrows():
    random_jitter = np.random.rand() / 5
    plt.text(x=row['T3PA'], y=(row['3P%']) + random_jitter, s=row['Player'])

for i, row in steph_curry.iterrows():
    random_jitter = np.random.rand() / 5
    plt.text(x=row['T3PA']*0.85, y=(row['3P%']*.9) + random_jitter, s=row['Player'])
```

```python
df_good_shooters = rs_df[f1 & f2]
df_good_shooters[['Player','3P','3PM','3PA','T3PA','3P%']].sort_values(by=['3P'], ascending=False).head(10)
```

# Most Efficient Team in terms of shooting

menghapus row tim yang tidak relevan dalam turnamen

```python
indexAge = rs_df[ (rs_df['Tm'] == 'TOT') ].index
rs_df =rs_df.drop(indexAge , inplace=False)
```

mengelompokkan setiap tim

```python
rs_df['eFG%'] = (rs_df['TFG'] + (.5*rs_df['3PM']))/rs_df['TFGA']
team_avg = rs_df.groupby(['Tm']).mean()
team_sum = rs_df.groupby(['Tm']).sum()
```

mencari efisiensi shooting (didapat dari sportingcharts.com)  
efisiensi diubah karena perhitungan masih per game bukan total dari points yang berhasil didapatkan (setiap pemain memiliki jumlah games berbeda)

```python
team_sum['eFG%'] = team_avg['eFG%']
team_sum['3P%'] = team_avg['3P%']
```

```python
sns.set()
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(data=rs_df, x="eFG%", y="Tm" , ax=ax , color='b',errorbar = None)
sns.barplot(data=rs_df, x="3P%", y="Tm" , ax=ax , color='r',errorbar = None)
ax.set(xlabel='Black-3Pperc. Grey-eFGperc.', ylabel='Teams')
ax.set_xlim(.2, .6)
```
